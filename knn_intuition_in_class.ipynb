```python
#features from our tain_data
feats = kdata[['variance', 'curtosis']].values
print(feats)
```

```python
#targets from our train data
target = kdata['class'].values
print(target)
```

```python
#importing numpy for our arrays
import numpy as np
```

```python
#new data item to be classified
new_data = np.array([2.37180, 0.015989])
new_data
```

```python
#define function for euclidian distance
def euc_dist(new_d, feat_d):
    m = feat_d.shape[0] #number of rows in feature data/ determine the dimensions of array
    distance = np.zeros(m) #initialize array of zeros
    for i in range(m):
        distance[i] = np.sqrt((new_d[0] - feat_d[i][0])**2 + (new_d[1] - feat_d[i][1])**2)
    return distance
```

```python
# find the distances of each item in our data
d = euc_dist(new_data, feats)
d
```

```python
#function to predict the target given k neighbors
def pred_knn(euc_d, targ, k):
    m = targ.shape[0] #number of rows in target data/dimensions of the array
    arr = np.zeros([m,2]) #initialize array of zeros with dimensions of m x 2 (2 columns)
    for i in range(m):
        arr[i] = np.array([euc_d[i], targ[i]])
        print(arr[i])
        sort_arr = np.sort(arr, axis=0)
        sort_arr = sort_arr[:k]
    return sort_arr
```

```python
first_k = pred_knn(d, target, 5) #predict the target given 5 neighbors
```
